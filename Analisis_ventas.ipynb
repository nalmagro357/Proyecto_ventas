```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# URL del archivo en Github
url= "https://raw.githubusercontent.com/nalmagro357/Proyecto_ventas/refs/heads/main/ventas_productos.csv"

# Cargar los datos del CSV
df = pd.read_csv(url)

# Calcular el precio total por producto
df["Precio_Total"] = df["Cantidad"] * df["Precio"]

# Crear el gráfico de barras con colores distintos en tonos de azul
plt.figure(figsize=(8,5))
colors = plt.cm.Blues(np.linspace(0.4, 1, len(df)))  # Generar diferentes tonos de azul

bars = plt.bar(df["Productos"], df["Precio_Total"], color=colors)

# Agregar etiquetas con los valores en la parte superior de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:.2f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Configuración de etiquetas y título
plt.xlabel("Producto")
plt.ylabel("Precio Total")
plt.title("Precio Total por Producto")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Guardar el gráfico como imagen PNG
plt.savefig("precio_total_productos.png")

# Mostrar el gráfico
plt.show()

# Guardar el DataFrame con los cálculos en un nuevo archivo CSV
df.to_csv("ventas_productos_con_precio_total.csv", index=False)

print("Análisis completado. Se generaron los archivos 'precio_total_productos.png' y 'ventas_productos_con_precio_total.csv'.")
```
